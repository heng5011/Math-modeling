```python
import numpy as np
import math
import csv


def yingshe(x, max_x, min_x) :
    return (x - min_x) / (max_x - min_x)

states = ['日本', '德国', '俄罗斯', '韩国', '美国', '南非', '巴西', '西班牙', '伊朗', '意大利', '印度', '中国',
         ]
infection = []
dead = []
cure = []
country = []

for state in states :
    g = open('F:\\下载\\建模\\东三省建模a\\Q_1\\欧式距离\\inputData.csv', 'r')
    reader = csv.reader(g)
    
    #dates = []
    #day = []
    ii = 0
    for row in reader:
        if (row[0] == state and ii > 0):
            infection.append(float(row[1]))
            # print(date_object,float(row[8]),float(row[7]),float(row[6]))
            dead.append(float(row[2]))
            cure.append(float(row[3]))
            country.append(row[0])
        ii += 1
    g.close()
    #print(ganran)
    #print(dead)
    #print(cure)
    #print(state)
infection_max = max(infection)
infection_min = min(infection)
dead_max = max(dead)
dead_min = min(dead)
cure_max = max(cure)
cure_min = min(cure)
#print(infection_max)
#print(infection_min)
#print(infection[0])

#感染
infection_set = []
dead_set = []
cure_set = []

for i in range(0, 12) :
    infection_set.append(yingshe(infection[i], infection_max, infection_min))
    dead_set.append(yingshe(dead[i], dead_max, dead_min))
    cure_set.append(yingshe(cure[i], cure_max, cure_min))

# print('感染映射:', infection_set)
# print('死亡映射:', dead_set)
# print('治愈映射:', cure_set)

# print(dead_set[11])
dis = []

#### 11个国家和中国对比
print('11个国家和中国对比的欧式距离')
for i in range(0, 11) :
    temp = (infection_set[i] - infection_set[11]) * (infection_set[i] - infection_set[11]) + (dead_set[i] - dead_set[11]) * (dead_set[i] - dead_set[11]) + (cure_set[i] - cure_set[11]) * (cure_set[i] - cure_set[11])
    dis.append(math.sqrt(temp))
    print(country[i], dis[i])
print(dis)
temp_1, temp_2, temp_3 = 0, 0, 0
for i in range(0, 12) :
    temp_1 += infection_set[i]
    temp_2 += dead_set[i]
    temp_3 += cure_set[i]
    average_infection = temp_1 / 12
    average_dead = temp_2 / 12
    average_cure = temp_3 / 12

print('\n-------------------------------\n 12个国家与平均值对比的欧式距离')
print('感染率平均值', average_infection)
print('死亡率平均值', average_dead)
print('治愈率平均值', average_cure)

#### 12个国家与平均值对比
for i in range(0, 12) :
    temp = (infection_set[i] - average_infection) * (infection_set[i] - average_infection) + (dead_set[i] - average_dead) * (dead_set[i] - average_dead) + (cure_set[i] - average_cure) * (cure_set[i] - average_cure)
    dis.append(math.sqrt(temp))
    print(country[i], dis[i])
print(dis)
```
