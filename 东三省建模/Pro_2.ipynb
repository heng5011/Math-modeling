```python
import numpy as np
A = np.array([[3,-1],[-1,3]])
print('打印A：\n{}'.format(A))
a, b = np.linalg.eig(A)
print('打印特征值a：\n{}'.format(a))
print('打印特征向量b：\n{}'.format(b))
```

```python
import numpy as np
A = np.array([[1, 5, 5 / 7, 5 / 3, 5 / 9], 
              [1 / 5, 1, 1 / 7, 1 / 3, 1 / 9],
              [7 / 5, 7, 1, 3 / 7, 1 / 3],
              [3 / 5, 3, 7 / 3, 1, 1 / 3],
              [9 / 5, 9, 3, 3, 1]
             ])
print(A)
a, b = np.linalg.eig(A)
print('打印特征值a：\n{}'.format(a))
print('打印特征向量b：\n{}'.format(b))
```
